```python
import pandas as pd
import numpy as np

# ==========================
# Load Dataset
# ==========================

file_path = r"E:\vs code programs\python\gridlock\Astram event data_anonymized - Astram event data_anonymizedb40ac87.csv"

df = pd.read_csv(file_path)

print("\nDataset Loaded Successfully")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

# ==========================
# Basic Information
# ==========================

print("\nDataset Info")
print(df.info())

print("\nFirst 5 Rows")
print(df.head())

# ==========================
# Column Names
# ==========================

print("\nColumns:")
for col in df.columns:
    print(col)

# ==========================
# Missing Values
# ==========================

print("\nMissing Values:")
missing = df.isnull().sum()

missing = missing[missing > 0]

print(missing.sort_values(ascending=False))

# ==========================
# Duplicate Records
# ==========================

duplicates = df.duplicated().sum()

print(f"\nDuplicate Rows: {duplicates}")

# ==========================
# Convert Datetime Columns
# ==========================

date_columns = [
    "start_datetime",
    "end_datetime",
    "resolved_datetime",
    "closed_datetime"
]

for col in date_columns:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")

print("\nDatetime Conversion Completed")

# ==========================
# Create Duration Feature
# ==========================

if (
    "start_datetime" in df.columns
    and "resolved_datetime" in df.columns
):

    df["duration_minutes"] = (
        df["resolved_datetime"]
        -
        df["start_datetime"]
    ).dt.total_seconds() / 60

    print("\nDuration Feature Created")

    print(df["duration_minutes"].describe())

# ==========================
# Extract Time Features
# ==========================

if "start_datetime" in df.columns:

    df["hour"] = df["start_datetime"].dt.hour
    df["day"] = df["start_datetime"].dt.day
    df["month"] = df["start_datetime"].dt.month
    df["weekday"] = df["start_datetime"].dt.day_name()

print("\nTime Features Created")

# ==========================
# Save Clean Dataset
# ==========================

df.to_csv("cleaned_event_data.csv", index=False)

print("\nCleaned Dataset Saved")
print("File: cleaned_event_data.csv")
```
