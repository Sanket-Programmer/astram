```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv("cleaned_event_data.csv")

print("Dataset Shape:", df.shape)

# ==============================
# Event Type Distribution
# ==============================

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="event_type",
    order=df["event_type"].value_counts().index
)

plt.title("Event Type Distribution")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ==============================
# Top Event Causes
# ==============================

plt.figure(figsize=(12,6))

top_causes = df["event_cause"].value_counts().head(10)

sns.barplot(
    x=top_causes.values,
    y=top_causes.index
)

plt.title("Top 10 Event Causes")
plt.tight_layout()
plt.show()

# ==============================
# Priority Distribution
# ==============================

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="priority",
    order=df["priority"].value_counts().index
)

plt.title("Priority Distribution")
plt.tight_layout()
plt.show()

# ==============================
# Road Closure Requirement
# ==============================

plt.figure(figsize=(6,6))

df["requires_road_closure"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Road Closure Requirement")
plt.ylabel("")
plt.show()

# ==============================
# Top Police Stations
# ==============================

plt.figure(figsize=(12,6))

top_ps = df["police_station"].value_counts().head(10)

sns.barplot(
    x=top_ps.values,
    y=top_ps.index
)

plt.title("Top 10 Police Stations by Event Count")
plt.tight_layout()
plt.show()

# ==============================
# Top Zones
# ==============================

zone_df = df[df["zone"].notna()]

plt.figure(figsize=(12,6))

top_zones = zone_df["zone"].value_counts().head(10)

sns.barplot(
    x=top_zones.values,
    y=top_zones.index
)

plt.title("Top Zones")
plt.tight_layout()
plt.show()

# ==============================
# Monthly Event Trend
# ==============================

if "month" in df.columns:

    monthly = df["month"].value_counts().sort_index()

    plt.figure(figsize=(10,5))

    sns.lineplot(
        x=monthly.index,
        y=monthly.values,
        marker="o"
    )

    plt.title("Monthly Event Trend")
    plt.xlabel("Month")
    plt.ylabel("Number of Events")
    plt.tight_layout()
    plt.show()

print("\nEDA Completed")
```
