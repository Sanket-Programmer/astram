```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# =====================
# Load Data
# =====================

df = pd.read_csv("cleaned_event_data.csv")

# =====================
# Keep Useful Columns
# =====================

features = [
    "event_type",
    "event_cause",
    "requires_road_closure",
    "police_station",
    "zone",
    "corridor",
    "latitude",
    "longitude",
    "priority"
]

df = df[features]

# =====================
# Fill Missing Values
# =====================

df["zone"] = df["zone"].fillna("Unknown")
df["corridor"] = df["corridor"].fillna("Unknown")

# =====================
# Encode Categories
# =====================

encoders = {}

cat_cols = [
    "event_type",
    "event_cause",
    "police_station",
    "zone",
    "corridor"
]

for col in cat_cols:

    le = LabelEncoder()

    df[col] = le.fit_transform(df[col])

    encoders[col] = le

# =====================
# Target Variable
# =====================

df["priority"] = df["priority"].map({
    "Low":0,
    "High":1
})

# Remove any unexpected values
df = df.dropna()

# =====================
# Save
# =====================

df.to_csv(
    "ml_dataset.csv",
    index=False
)

print("ML Dataset Created")
print(df.shape)

print(df.head())
```
